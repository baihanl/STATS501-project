```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Replace these with your actual values
metrics = ['AUC', 'BER', 'F1', 'Accuracy']
models = ['Logistics Regression', 'GAM', 'Random Forest', 'SVM', 'XGBoost']

# Sample values for each metric for each model
# Each column represents a model and each row a metric
values = np.array([
    [0.834, 0.850, 0.831, 0.818, 0.818],   # AUC values for each model
    [0.277, 0.280, 0.253, 0.326, 0.273],  # BER values for each model
    [0.611, 0.633, 0.611, 0.519, 0.587],   # F1 values for each model
    [0.785, 0.764, 0.775, 0.797, 0.7661]   # Accuracy values for each model
])

n_metrics = len(metrics)
n_models = len(models)
bar_width = 0.15
index = np.arange(n_metrics)

# Creating a bar plot
fig, ax = plt.subplots(figsize=(10, 6))

for i in range(n_models):
    ax.bar(index + i*bar_width, values[:, i], bar_width, label=models[i])

# Setting the x axis labels and chart title
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Comparison of Model Performance Across Various Metrics')
ax.set_xticks(index + bar_width * (n_models-1) / 2)
ax.set_xticklabels(metrics)
ax.legend()

plt.show()
```
